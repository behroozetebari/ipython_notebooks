# MODICE area by country, 2000-2014

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%pylab inline
```

```python
filename = 'modice_v4_3strikes_by_country_by_yr.txt'
df = pd.read_csv( filename, delim_whitespace=True, index_col=0 )
df
```

```python
df = df[['Tajikistan','Kyrgyz_Republic','Kazakhstan','Uzbekistan','Turkmenistan']]
df
```

```python
ax = df.plot(style='-o')
ax.legend(bbox_to_anchor=(1.5,1.0))
ax.set(title="MODICE by Country", ylabel='MODICE area ($km^2$)' )
```
