# MODICE v04 area by country, 2000-2014

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%pylab inline
```

```python
filename = 'modice_v4_3strikes_by_country_by_yr.txt'
df = pd.read_csv( filename, delim_whitespace=True, index_col=0 )
df
```

## Reorder columns, descending area

```python
df = df[['Tajikistan','Kyrgyz_Republic','Kazakhstan','Uzbekistan','Turkmenistan']]
df
```

```python
my_colors = list(['b','g','r','c','m'])
ax = df.plot(style='-o', color=my_colors)
ax.legend(bbox_to_anchor=(1.5,1.0))
ax.set(title="MODICE by Country", ylabel='MODICE area ($km^2$)' )
```

## Just look at detail on three lowest lines:

```python
subdf = df[['Kazakhstan','Uzbekistan','Turkmenistan']]
subdf
```

```python
ax2 = subdf.plot(style='-o', color=my_colors[2:] )
ax2.legend(bbox_to_anchor=(1.5,1.0))
ax2.set(title="MODICE by Country", ylabel='MODICE area ($km^2$)' )
```
